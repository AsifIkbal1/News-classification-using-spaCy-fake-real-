```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

## Get data (CSV dataframe)

```python
df1 = pd.read_csv('../input/fake-and-real-news-dataset/Fake.csv')
df2 = pd.read_csv('../input/fake-and-real-news-dataset/True.csv')
df1['label'] = 0
df2['label'] = 1
df = pd.concat([df1, df2])

df
```

## Add news field (Vector of text) in to dataframe

```python
doc = nlp('Donald Trump Sends Out Embarrassing New Year')
doc.vector.shape
```

```python
df['vector'] = df['text'].apply(lambda text: nlp(text).vector)
```

```python
df
```

## Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.vector.values,
    df.label,
    test_size=0.2,
    random_state=42
)
```

## Convert to 2d array

```python
X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

## Scale data and create model

```python
scaler = MinMaxScaler()
scaled_train_embed = scaler.fit_transform(X_train_2d)
scaled_test_embed = scaler.transform(X_test_2d)


clf = MultinomialNB()
clf.fit(scaled_train_embed, y_train)
```

## Classification report

```python
y_pred = clf.predict(scaled_test_embed)

print(classification_report(y_test, y_pred))
```
